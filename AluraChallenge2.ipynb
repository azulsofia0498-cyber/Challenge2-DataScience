```python
import pandas as pd

duplicados = df_relevante.duplicated().sum()
print(f"Filas duplicadas encontradas: {duplicados}")
df_relevante = df_relevante.drop_duplicates()

num_cols = ["customer.tenure", "account.Charges.Monthly", "account.Charges.Total"]
for col in num_cols:
    df_relevante[col] = pd.to_numeric(df_relevante[col], errors='coerce')

# Opciones: reemplazar NaN en cargos con 0, en tenure con mediana, en categorías con 'Unknown'
df_relevante["customer.tenure"].fillna(df_relevante["customer.tenure"].median(), inplace=True)
df_relevante["account.Charges.Monthly"].fillna(0, inplace=True)
df_relevante["account.Charges.Total"].fillna(0, inplace=True)

categoricas = [
    "customer.gender", "customer.SeniorCitizen", "customer.Partner", "customer.Dependents",
    "phone.PhoneService", "phone.MultipleLines",
    "internet.InternetService", "internet.OnlineSecurity", "internet.OnlineBackup",
    "internet.DeviceProtection", "internet.TechSupport", "internet.StreamingTV", "internet.StreamingMovies",
    "account.Contract", "account.PaperlessBilling", "account.PaymentMethod", "Churn"
]

for col in categoricas:
    df_relevante[col].fillna("Unknown", inplace=True)

# Ejemplo: unificar "No internet service" y "No phone service" a "No"
reemplazos = {
    "No internet service": "No",
    "No phone service": "No",
    "Yes": "Yes",  # solo para asegurar consistencia
    "No": "No"
}

cols_servicios = [
    "phone.PhoneService", "phone.MultipleLines",
    "internet.InternetService", "internet.OnlineSecurity", "internet.OnlineBackup",
    "internet.DeviceProtection", "internet.TechSupport", "internet.StreamingTV", "internet.StreamingMovies"
]

for col in cols_servicios:
    df_relevante[col] = df_relevante[col].replace(reemplazos)

print("\n=== Revisión final ===")
print(df_relevante.info())
print("\nValores únicos por categoría (ejemplo de un par de columnas):")
print("Churn:", df_relevante["Churn"].unique())
print("InternetService:", df_relevante["internet.InternetService"].unique())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.figure(figsize=(12,5))

# Conteo de clientes por Churn
churn_counts = df_relevante["Churn"].value_counts()

plt.subplot(1,2,1)
sns.barplot(x=churn_counts.index, y=churn_counts.values, color="#5DADE2")
plt.title("Número de clientes por Churn")
plt.ylabel("Cantidad de clientes")
plt.xlabel("Churn")
for i, v in enumerate(churn_counts.values):
    plt.text(i, v + 5, str(v), ha='center', fontweight='bold')


plt.subplot(1,2,2)
plt.pie(churn_counts.values, labels=churn_counts.index, autopct="%1.1f%%", colors=["#5DADE2","#F1948A"], startangle=90)
plt.title("Proporción de clientes que permanecen vs se dan de baja")
plt.axis('equal')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.set_style("whitegrid")
plt.figure(figsize=(18,20))

categoricas_churn = [
    "customer.gender", "customer.SeniorCitizen", "customer.Partner", "customer.Dependents",
    "phone.PhoneService", "phone.MultipleLines",
    "internet.InternetService", "internet.OnlineSecurity", "internet.OnlineBackup",
    "internet.DeviceProtection", "internet.TechSupport", "internet.StreamingTV", "internet.StreamingMovies",
    "account.Contract", "account.PaperlessBilling", "account.PaymentMethod"
]

n_cols = 3
n_rows = (len(categoricas_churn) + n_cols - 1) // n_cols
fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows*4))
axes = axes.flatten()

for i, col in enumerate(categoricas_churn):
    sns.countplot(data=df_relevante, x=col, hue="Churn", palette=["#5DADE2","#F1948A"], ax=axes[i])
    axes[i].set_title(f"Churn por {col}")
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Cantidad de clientes")
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].legend(title="Churn")

for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
df_grouped = df_relevante.groupby(['customer.tenure', 'Churn']).size().unstack(fill_value=0)

df_grouped['churn_rate'] = (df_grouped.get('Yes', 0) / df_grouped.sum(axis=1)) * 100

print("=== Tasa de churn por tiempo de contrato (meses) ===")
print(df_grouped['churn_rate'].map(lambda x: f'{x:.1f}%').to_string())
```

```python
import plotly.express as px

df_grouped = df_relevante.groupby(['customer.tenure', 'Churn']).size().unstack(fill_value=0)

df_grouped['churn_rate'] = (df_grouped.get('Yes', 0) / df_grouped.sum(axis=1)) * 100

fig = px.line(
    df_grouped.reset_index(),
    x='customer.tenure',
    y='churn_rate',
    markers=True,
    title='Tasa de Churn (%) por tiempo de permanencia',
    color_discrete_sequence=['red']
)

fig.update_layout(title_x=0.5)
fig.update_traces(hovertemplate='Tiempo de permanencia: %{x} mesesTasa de Churn: %{y:.1f}%')
fig.update_xaxes(title_text='Tiempo de permanencia (meses)')
fig.update_yaxes(title_text='Tasa de Churn (%)')

fig.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.figure(figsize=(18,5))

numericas = ["customer.tenure", "account.Charges.Monthly", "account.Charges.Total"]
titles = ["Antigüedad (tenure)", "Cargos Mensuales (Monthly Charges)", "Total Gastado (Total Charges)"]

for i, col in enumerate(numericas):
    plt.subplot(1,3,i+1)
    sns.boxplot(x="Churn", y=col, data=df_relevante, palette=["#5DADE2","#F1948A"])
    plt.title(f"{titles[i]} por Churn")
    plt.xlabel("Churn")
    plt.ylabel(titles[i])

plt.tight_layout()
plt.show()
```

En cambio, tarifas altas, falta de servicios complementarios y pagos manuales aumentan la probabilidad de cancelación
